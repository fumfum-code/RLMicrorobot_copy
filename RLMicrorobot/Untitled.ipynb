```python
def sygmoid(x,a,b):
    return 1.5/(1+np.exp((-x+a)/b)) -0.5

x = np.linspace(-10,10,100)
fig, ax = plt.subplots()

ax.set_xlim([-10, 10])
ax.set_ylim([-1.1, 1.1])
ax.plot(x,sygmoid(x,1,1.5))

ax.grid()

```
